function to tokenize the text data using automodel

```python
def tokenize_function(description):
  return tokenizer(description['text'],padding='max_length',truncation=True)
```

```python
tokenized_datasets=dataset.map(tokenize_function,batched=True)
```

```python
tokenized_datasets['train']
```

Consider a sample of tokenized data for training and evaluation

```python
small_train=tokenized_datasets['train'].shuffle(seed=42).select(range(1000))
small_eval=tokenized_datasets['train'].shuffle(seed=42).select(range(1000))
```

```python
small_train
```

```python
model=AutoModelForSequenceClassification.from_pretrained('bert-base-cased',num_labels=5)
```

```python
# dataset['train']
```

Training hyperparameters using trainer class

```python
from transformers import TrainingArguments,Trainer
training_args=TrainingArguments(output_dir='training_args_directory')
```

Metrics

```python
import numpy as np
from datasets import load_metric
metric = load_metric("accuracy")
```

```python
def compute_metric(eval_preds):
  logits,eval=eval_preds
  predictions=np.argmax(logits,axis=-1)
  return metric.compute(predictions=predictions,references=eval)
  
```

```python
training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

```python
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train,
    eval_dataset=small_eval,
    compute_metrics=compute_metric,
    )
```

Trainer and training the model

```python
trainer.train()
```

Pytorch way to train the model

```python
from torch.utils.data import DataLoader
train_dl=DataLoader(small_train,shuffle=True,batch_size=8)
eval_dl=DataLoader(small_eval,batch_size=8)
```

Optimizer

```python
from torch.optim import AdamW
optimizer=AdamW(model.parameters(),lr=5e-5)
```

```python
from transformers import get_scheduler
num_epochs=3
num_training_steps=num_epochs*(len(train_dl))
lr_scheduler=get_scheduler(name='linear',optimizer=optimizer,num_training_steps=num_training_steps,num_warmup_steps=0)
```

```python
device=torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
```

```python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()


for epoch in range(num_epochs):
  for batch in train_dl:
    batch={k: v.to(device) for k,v in batch.items()}
    outputs=model(**batch)
    loss=outputs.loss
    loss.backward()

    optimizer.step()
    lr_scheduler.step()
    optimizer.zero_grad()
    progress_bar.update(1)

    
```
